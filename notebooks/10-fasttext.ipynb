```python
# import as libraries importantes
import pandas as pd
import sys
from fasttext import load_model
```

```python
classifier = load_model("../data/commits.bin") 
df = pd.read_csv('../data/commits_processed.csv',  lineterminator='\n', encoding="utf-8")
df = df.replace('\n','', regex=True)
```

```python
# a coluna que queremos predizer
df.message
```

```python
# guarda a lista com as mensagens dos commits
commits = list(df['message'].astype(str))
```

```python
# prediz utilizando o modelo do fasttext
labels = classifier.predict(commits)  
```

```python
# processa os resultados
res = list(zip(*labels))
res_list = [x[0] for x in res]
lst2 = [item[0] for item in res_list]
```

```python
df['labels_predicted'] = lst2
```

```python
prediction = df[['message','labels_predicted']]
```
